# Pet Classification Model Using CNN. 

```python
import tensorflow as tf
```

```python
import warnings
warnings.filterwarnings('ignore')
```

Use Image Generators
Initialize
For Training Data and for Validating data

```python
train_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.)
validation_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255.)
```

Pass set of images to the generator to generate compatible images for CNN

```python
trainImagedata = train_generator.flow_from_directory('C:/Users/Saransh/Desktop/Class PPT/DL project/data/train',batch_size = 32,class_mode='binary',target_size=(64,64))
validationImagedata = train_generator.flow_from_directory('C:/Users/Saransh/Desktop/Class PPT/DL project/data/test',batch_size = 32,class_mode='binary',target_size=(64,64))
```

Architecting CNN Model

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Conv2D(32,(3,3),input_shape=(64,64,3),activation='relu',padding='same',strides=(1,1)))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))


model.add(tf.keras.layers.Conv2D(16,(3,3),input_shape=(64,64,3),activation='relu',padding='same',strides=(1,1)))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(tf.keras.layers.Flatten())

model.add(tf.keras.layers.Dense(units=512,activation='relu'))
model.add(tf.keras.layers.Dense(units=256,activation='relu'))

model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

Compiling

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```
